```python
#  by tutedude lecture 
from selenium import webdriver 
from time import sleep
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
import time 
driver = webdriver.Edge()
driver.get('https://www.facebook.com/login/')
time.sleep(1)
driver.find_element(By.NAME,"email").send_keys('geekmicrosoft254@gmail.com')
driver.find_element(By.NAME,"pass").send_keys('xswxsw')
time.sleep(1)
driver.find_element(By.XPATH,'//*[@id="loginbutton"]').click()
time.sleep(1)
driver.get('https://www.facebook.com/home.php')
time.sleep(1)
p =driver.find_element(By.XPATH,"//*[@name='xhpc_message]")
time.sleep(1)
p.send_keys('Hi there!')
time.sleep(1)
buttons = driver.find_elements(By.TAG_NAME,'button')
time.sleep(2)

for button in buttons:
    if button.text == 'Post':
        button.click()
driver.quit()

```
