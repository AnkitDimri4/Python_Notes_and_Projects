#### **Search & Match Methods**

###### **Match Methods**

```python
import re # re for the module of regular expression 
string = "abc" # it match the pattern right start of the string
pattern = "a"
if re.match(pattern,string):
    print('Match Found')
else:
    print('Match not found')
```

###### **Search Methods**

```python
import re # re for the module of regular expression 
string = "vdfabc_t"
pattern = "_t"
if re.search(pattern,string):
    print('Match Found')
else:
    print('Match not found')
```

#### **Star Meta Character**

```python
import re # re for the module of regular expression 
string = "abbc"
pattern = "ab*c"
if re.search(pattern,string):
    print('Match Found')
else:
    print('Match not found')
```

```python
import re # re for the module of regular expression 
string = "abbc"
pattern = "ab+c"
if re.search(pattern,string):
    print('Match Found')
else:
    print('Match not found')
```

#### **Curly Braces Examples**

```python
import re # re for the module of regular expression 
string = "abbccdd"
pattern = "ab{2}c{2}"
if re.search(pattern,string):
    print('Match Found')
else:
    print('Match not found')
```

#### **Wildcard Example**

```python
import re
string = "abhbccdd" # Match not found
string = "abbccdd" # Martch found 
pattern = r"a.b"
if re.search(pattern,string):
    print('Match Found')
else:
    print('Match not found')
```

__________________________
#### **Optional Meta Character**

```python
import re
string = "python-file" 
pattern = r"python-?file" 
if re.search(pattern,string):
    print('Match Found')
else:
    print('Match not found')
```

__________________________
#### **Caret Meta Character**

```python
import re
# match must start this particular point
string = "9186566655"  
pattern = r"^91" 
if re.search(pattern,string):
    print('Match Found')
else:
    print('Match not found')
```

#### **Character classes part 1**

```python
import re
# We have more then five digits here  for the alphabet and character 
string = "a343" # all the string file must be contain more then 5 character  
string = "a3435555" # match found 
pattern = r"\d{5}" 
stringg = "u5"
patternn = r"\w{4}" # w for all the alpha numerical character

if re.match(patternn,stringg):
    print('Match Found')
else:
    print('Match not found')
```

#### **Character classes Part 2**

```python
import re
string = "Python"
pattern = r"^[a-zA-Z]ython" # ^[a-z] ^[A-Z] ^[Py]
if re.search(pattern,string):
    print('Match Found')
else:
    print('Match not found')
```
